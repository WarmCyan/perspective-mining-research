```python
flattened_pos = []
for pos_sentence in pos_sentences:
    sentence = []
    for pos in pos_sentence:
        sentence.append(pos[0])
    flattened_pos.append(sentence)
```

https://radimrehurek.com/gensim/models/ldamodel.html

```python
from gensim import corpora, models

dictionary = corpora.Dictionary([aspect_data.keys()])
corpus = [dictionary.doc2bow(sentence) for sentence in flattened_pos]
```

```python
import logging
logging.basicConfig(level=logging.INFO)

# note that in the paper, they mentioned 3000 iterations, is that the same as passes?
ldamodel = gensim.models.LdaMulticore(corpus, num_topics=20, id2word=dictionary, passes=20)
print(ldamodel.print_topics(num_topics=20, num_words=10))
```
