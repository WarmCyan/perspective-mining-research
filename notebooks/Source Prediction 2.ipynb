## TFIDF predictions

```python
import pandas as pd
#with open("../data/cache/testsentimentanalysis_partial/tfidif.json") as in_file:
    #tfidf_vectors = json.load(in_file)
tfidf_df = pd.read_json("../data/cache/testsentimentanalysis_partial/tfidif.json")
tfidf_df
```

```python
from sklearn.model_selection import train_test_split

X = tfidf_df
y = articles.source

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=42, multi_class='ovr') # one vs all
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

## ASVec

```python
as_df = pd.read_json("../data/cache/testsentimentanalysis_partial/as_vectors/doc_as_vectors.json")
as_df
```

```python

X = as_df
y = articles.source

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=42, multi_class='ovr') # one vs all
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train,y_train)
```

```python
clf.score(X_test, y_test)
```
