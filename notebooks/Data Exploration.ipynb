```python
import pandas as pd

def load_data():
    article_table1 = pd.read_csv("../data/raw/kaggle1/articles1.csv")
    article_table2 = pd.read_csv("../data/raw/kaggle1/articles2.csv")
    article_table3 = pd.read_csv("../data/raw/kaggle1/articles3.csv")
    article_table = pd.concat([article_table1, article_table2, article_table3])
    return article_table
```

```python
articles = load_data()
articles
```

```python
articles.publication.value_counts()
```

```python
trump_articles = articles[articles.content.str.contains("Trump")]
print(trump_articles.shape)
trump_articles.publication.value_counts()
```
