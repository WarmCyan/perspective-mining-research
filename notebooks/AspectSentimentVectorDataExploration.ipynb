# Article Exploration

```python
articles.iloc[0].text
```

```python
aspect_sentiments = as_df.iloc[0]
```

```python
sentiment_indices = {}
for index, val in aspect_sentiments.iteritems():
    if val != 0.0:
        sentiment_indices[index] = val

sentiment_indices
```

```python
from collections import OrderedDict

num_aspects = len(list(aspect_data.keys()))

labeled_sentiments = OrderedDict()

for key in sentiment_indices.keys():    
    neg = False
    obj = False
    
    aspect_key = key
    
    if key >= num_aspects and key < num_aspects*2:
        neg = True
        aspect_key = key - num_aspects
    elif key >= num_aspects*2:
        obj = True
        aspect_key = key - num_aspects*2
    
    label = list(aspect_data.keys())[aspect_key]
    if neg: label += "_neg"
    if obj: label += "_obj"
    value = sentiment_indices[key]
    labeled_sentiments[label] = value
    
labeled_sentiments = OrderedDict(sorted(labeled_sentiments.items(), key=lambda x:x[1]))
labeled_sentiments
```

# Find opposing perspectives

```python
as_df.iloc[0][as_df.iloc[0] != 0.0].index
```

(make aspect presence vectors)

```python
presence_df = as_df.copy()
```

```python
# turn all non-zero's into ones
def oneify(x):
    if x != 0.0:
        return 1.0
    return 0.0

presence_df = presence_df.applymap(oneify)
```

```python
presence_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cs_p = cosine_similarity(presence_df)
cs_p_df = pd.DataFrame(cs_p)
```

```python
sim = cs_p_df.loc[0]
sim[(sim < .99) & (sim > .25)].sort_values(ascending=False)
```

```python
closest = sim[(sim < .99) & (sim > .25)]
```

```python
closest_indices = list(closest.index)
closest_indices.insert(0, 0)
closest_indices
```

```python
similar_as_df = as_df.iloc[closest_indices]
similar_as_df
```

```python
test = pd.DataFrame(cosine_similarity(similar_as_df))
test.index = similar_as_df.index
test
```

```python
test.sort_values(by=[0]).index[0]
```
