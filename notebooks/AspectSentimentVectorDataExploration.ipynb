# AS_Vec

```python
as_df = pd.read_json("../data/cache/testsentimentanalysis_partial/as_vectors/doc_as_vectors.json")
as_df
```

```python
as_df.describe()
```

```python
#as_df.idxmax
print(as_df.max().max())
print(as_df.max().idxmax())
```

```python
as_df[1725].max()
```

```python
print(as_df.min().min())
print(as_df.min().idxmin())
```

try to rank rows by their abs value of min/max

```python
maximum_sentiment = as_df.loc[:, as_df.max().sort_values(ascending=False).index]
maximum_sentiment.max()
```

```python
minimum_sentiment = as_df.loc[:, as_df.min().sort_values(ascending=True).index]
minimum_sentiment.min()
```

Try to find maximum distance between maximum and minimum

```python
(as_df.max() - as_df.min()).max()
```

```python
distance_sentiment = as_df.loc[:, (as_df.max() - as_df.min()).sort_values(ascending=False).index]
distance_sentiment.describe()
```

How many aspects have 0 sentiment?

```python
((as_df.max() - as_df.min()) == 5).value_counts()
```
